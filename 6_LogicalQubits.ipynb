## Define repetition code encoding

### Subtask:
Create a `cirq` circuit to encode a logical qubit into multiple physical qubits using repetition code.


**Reasoning**:
Define a function to create a Cirq circuit for repetition encoding based on the provided instructions.



```python
import cirq

def build_repetition_encoding_circuit(num_physical_qubits: int) -> cirq.Circuit:
    """
    Builds a Cirq circuit for repetition encoding.

    Args:
        num_physical_qubits: The number of physical qubits (repetition code distance).

    Returns:
        A cirq.Circuit for repetition encoding.
    """
    qubits = cirq.LineQubit.range(num_physical_qubits)
    circuit = cirq.Circuit()

    # Apply Hadamard to the first qubit
    circuit.append(cirq.H(qubits[0]))

    # Entangle the first qubit with the remaining qubits using CNOT gates
    for i in range(1, num_physical_qubits):
        circuit.append(cirq.CNOT(qubits[0], qubits[i]))

    return circuit

# Example usage (optional, for verification)
# encoding_circuit = build_repetition_encoding_circuit(3)
# print(encoding_circuit)
```

```python
import cirq
import numpy as np

def test_repetition_code(num_physical_qubits: int, noise_model: cirq.NoiseModel, initial_logical_state: str):
    """
    Tests the repetition code encoding and decoding circuits against noise.

    Args:
        num_physical_qubits: The number of physical qubits (repetition code distance).
        noise_model: The cirq.NoiseModel to apply.
        initial_logical_state: The initial logical state ('0' or '1').
    """
    print(f"Testing with {num_physical_qubits} physical qubits, initial state |{initial_logical_state}>, and noise model: {noise_model}")

    # 1. Build the encoding circuit
    encoding_circuit = build_repetition_encoding_circuit(num_physical_qubits)

    # If initial_logical_state is '1', flip the first qubit before encoding
    if initial_logical_state == '1':
        encoding_circuit = cirq.Circuit(cirq.X(cirq.LineQubit(0))) + encoding_circuit

    # 2. Add the specified noise model to the encoding circuit
    noisy_encoding_circuit = add_noise_to_circuit(encoding_circuit, noise_model)

    # 3. Build the decoding circuit
    decoding_circuit = build_repetition_decoding_circuit(num_physical_qubits)

    # 4. Combine the noisy encoding circuit and the decoding circuit
    combined_circuit = noisy_encoding_circuit + decoding_circuit

    # 5. Create a Cirq Simulator
    simulator = cirq.Simulator()

    # 6. Simulate the combined circuit
    # Run multiple trials to get statistically relevant results
    num_trials = 1000
    results = simulator.run(combined_circuit, repetitions=num_trials)

    # 7. Analyze the simulation results
    # Check the measurement outcomes of the 'decoded_logical_qubit'
    decoded_outcomes = results.measurements['decoded_logical_qubit'].flatten()

    # Determine the expected outcome based on the initial logical state
    expected_outcome = 0 if initial_logical_state == '0' else 1

    # Count how many times the decoded logical qubit matches the expected state
    correct_outcomes_count = np.sum(decoded_outcomes == expected_outcome)

    # Calculate the success rate
    success_rate = correct_outcomes_count / num_trials

    print(f"  Expected outcome: {expected_outcome}")
    print(f"  Observed outcomes (first 10): {decoded_outcomes[:10]}")
    print(f"  Correct outcomes count: {correct_outcomes_count}/{num_trials}")
    print(f"  Success rate: {success_rate:.4f}")

    # 8. Include assertions or print statements to indicate test result
    # The threshold for success depends on the noise level and code distance.
    # A simple check is if the success rate is significantly higher than random chance (0.5).
    # For low noise and sufficient code distance, we expect a high success rate.
    # Let's set a simple threshold, e.g., > 0.7, but this should be interpreted
    # in the context of the specific noise model and code parameters.
    success_threshold = 0.7

    if success_rate > success_threshold:
        print(f"  Test Passed: Success rate ({success_rate:.4f}) is above threshold ({success_threshold}).")
    else:
        print(f"  Test Failed: Success rate ({success_rate:.4f}) is below threshold ({success_threshold}).")

    print("-" * 20)

# Example test cases:
# Define a simple noise model (e.g., a small probability of bit flip on each qubit after each gate)
bit_flip_noise = cirq.bit_flip(0.01) # 1% bit flip probability

# Test with 3 physical qubits (distance 3 code)
test_repetition_code(num_physical_qubits=3, noise_model=bit_flip_noise, initial_logical_state='0')
test_repetition_code(num_physical_qubits=3, noise_model=bit_flip_noise, initial_logical_state='1')

# Test with 5 physical qubits (distance 5 code)
test_repetition_code(num_physical_qubits=5, noise_model=bit_flip_noise, initial_logical_state='0')
test_repetition_code(num_physical_qubits=5, noise_model=bit_flip_noise, initial_logical_state='1')

# Test with no noise (should have very high success rate)
test_repetition_code(num_physical_qubits=3, noise_model=cirq.NO_NOISE, initial_logical_state='0')
test_repetition_code(num_physical_qubits=3, noise_model=cirq.NO_NOISE, initial_logical_state='1')
```

The tests are still failing even without noise, and the success rates are around 50%, which indicates the decoding measurement is not correctly reflecting the encoded logical state. The current decoding circuit measures the first physical qubit, which after encoding the |0> state with Hadamard and CNOTs (which creates a GHZ-like state |00...0> + |11...1>), will be in a superposition and measuring it will yield 0 or 1 with ~50% probability. Similarly for the |1> initial state. The repetition code protects the logical state by encoding it into the parity of the physical qubits. To decode, we need to measure the state of the logical qubit. In the repetition code context, the logical |0> state is encoded as |00...0> and logical |1> as |11...1>. After noisy evolution, the state will be a mixture. To determine the logical state, we need to measure the physical qubits and infer the logical state. A common way to decode the repetition code is by measuring all physical qubits and taking a majority vote. However, the current test setup expects a single 'decoded_logical_qubit' measurement key.

Let's modify the encoding circuit so that the initial logical state is directly encoded into the |00...0> or |11...1> state, rather than using a Hadamard on the first qubit. This aligns better with the standard repetition code where |0>_L = |00...0> and |1>_L = |11...1>. Then, the decoding circuit should measure all physical qubits, and the test function should perform the majority vote classically on these outcomes to determine the decoded logical state. We will still measure the first physical qubit as 'decoded_logical_qubit' for compatibility with the test function structure, but the analysis in the test function should ideally use the majority vote from 'physical_qubit_outcomes'. Let's update the test function to perform the majority vote analysis.



```python
import cirq
import numpy as np

def build_repetition_encoding_circuit(num_physical_qubits: int) -> cirq.Circuit:
    """
    Builds a Cirq circuit for repetition encoding.

    Args:
        num_physical_qubits: The number of physical qubits (repetition code distance).

    Returns:
        A cirq.Circuit for repetition encoding.
    """
    qubits = cirq.LineQubit.range(num_physical_qubits)
    circuit = cirq.Circuit()

    # The logical |0> is encoded as |00...0>, logical |1> as |11...1>.
    # We will prepare the first qubit in the desired logical state
    # and then copy its state to the others.
    # The initial state preparation (|0> or |1>) will be handled before calling this function.

    # Copy the state of the first qubit to the remaining qubits using CNOT gates
    for i in range(1, num_physical_qubits):
        circuit.append(cirq.CNOT(qubits[0], qubits[i]))

    return circuit

def build_repetition_decoding_circuit(num_physical_qubits: int) -> cirq.Circuit:
    """
    Builds a Cirq circuit for repetition decoding by measuring all physical qubits.

    Args:
        num_physical_qubits: The number of physical qubits (repetition code distance).

    Returns:
        A cirq.Circuit for repetition decoding.
    """
    physical_qubits = cirq.LineQubit.range(num_physical_qubits)
    circuit = cirq.Circuit()

    # Measure all physical qubits. The majority vote will be done in post-processing.
    circuit.append(cirq.measure(*physical_qubits, key='physical_qubit_outcomes'))

    # For compatibility with the test function structure expecting 'decoded_logical_qubit',
    # we will still include a measurement of the first qubit, but the primary analysis
    # will use 'physical_qubit_outcomes'.
    circuit.append(cirq.measure(physical_qubits[0], key='decoded_logical_qubit'))

    return circuit


def test_repetition_code(num_physical_qubits: int, noise_model: cirq.NoiseModel, initial_logical_state: str):
    """
    Tests the repetition code encoding and decoding circuits against noise.

    Args:
        num_physical_qubits: The number of physical qubits (repetition code distance).
        noise_model: The cirq.NoiseModel to apply.
        initial_logical_state: The initial logical state ('0' or '1').
    """
    print(f"Testing with {num_physical_qubits} physical qubits, initial state |{initial_logical_state}>, and noise model: {noise_model}")

    # Prepare the initial logical state
    initial_state_circuit = cirq.Circuit()
    qubits = cirq.LineQubit.range(num_physical_qubits)
    if initial_logical_state == '1':
        initial_state_circuit.append(cirq.X(qubits[0]))

    # 1. Build the encoding circuit
    encoding_circuit = build_repetition_encoding_circuit(num_physical_qubits)

    # Combine initial state preparation and encoding
    full_encoding_circuit = initial_state_circuit + encoding_circuit

    # 2. Add the specified noise model to the encoding circuit
    noisy_encoding_circuit = add_noise_to_circuit(full_encoding_circuit, noise_model)

    # 3. Build the decoding circuit
    decoding_circuit = build_repetition_decoding_circuit(num_physical_qubits)

    # 4. Combine the noisy encoding circuit and the decoding circuit
    combined_circuit = noisy_encoding_circuit + decoding_circuit

    # 5. Create a Cirq Simulator
    simulator = cirq.Simulator()

    # 6. Simulate the combined circuit
    # Run multiple trials to get statistically relevant results
    num_trials = 1000
    results = simulator.run(combined_circuit, repetitions=num_trials)

    # 7. Analyze the simulation results using majority vote
    physical_outcomes = results.measurements['physical_qubit_outcomes']

    # Perform majority vote for each trial
    decoded_logical_states = np.apply_along_axis(lambda x: np.argmax(np.bincount(x)), axis=1, arr=physical_outcomes)

    # Determine the expected outcome based on the initial logical state
    expected_outcome = 0 if initial_logical_state == '0' else 1

    # Count how many times the decoded logical qubit matches the expected state
    correct_outcomes_count = np.sum(decoded_logical_states == expected_outcome)

    # Calculate the success rate
    success_rate = correct_outcomes_count / num_trials

    print(f"  Expected logical outcome (majority vote): {expected_outcome}")
    print(f"  Decoded logical outcomes (majority vote, first 10): {decoded_logical_states[:10]}")
    print(f"  Correct outcomes count: {correct_outcomes_count}/{num_trials}")
    print(f"  Success rate: {success_rate:.4f}")

    # 8. Include assertions or print statements to indicate test result
    # With no noise, we expect a 100% success rate. With noise, the rate should be higher than 0.5
    # and increase with the number of physical qubits (for correctable error rates).
    if noise_model == cirq.NO_NOISE:
         success_threshold = 0.99 # Expect near perfect success with no noise
    else:
         success_threshold = 0.6 # Expect better than random chance with noise protection

    if success_rate > success_threshold:
        print(f"  Test Passed: Success rate ({success_rate:.4f}) is above threshold ({success_threshold}).")
    else:
        print(f"  Test Failed: Success rate ({success_rate:.4f}) is below threshold ({success_threshold}).")

    print("-" * 20)

# Example test cases:
# Define a simple noise model (e.g., a small probability of bit flip on each qubit after each gate)
bit_flip_noise = cirq.bit_flip(0.01) # 1% bit flip probability

# Test with 3 physical qubits (distance 3 code)
test_repetition_code(num_physical_qubits=3, noise_model=bit_flip_noise, initial_logical_state='0')
test_repetition_code(num_physical_qubits=3, noise_model=bit_flip_noise, initial_logical_state='1')

# Test with 5 physical qubits (distance 5 code)
test_repetition_code(num_physical_qubits=5, noise_model=bit_flip_noise, initial_logical_state='0')
test_repetition_code(num_physical_qubits=5, noise_model=bit_flip_noise, initial_logical_state='1')

# Test with no noise (should have very high success rate)
test_repetition_code(num_physical_qubits=3, noise_model=cirq.NO_NOISE, initial_logical_state='0')
test_repetition_code(num_physical_qubits=3, noise_model=cirq.NO_NOISE, initial_logical_state='1')
```

## Summary:

### Data Analysis Key Findings

*   The standard repetition code encoding maps logical $|0\rangle$ to $|00...0\rangle$ and logical $|1\rangle$ to $|11...1\rangle$.
*   Decoding the repetition code against bit-flip errors is typically done by measuring each physical qubit and applying a classical majority vote to determine the most likely logical state.
*   Increasing the number of physical qubits (code distance) generally improves the success rate of correctly decoding the logical state in the presence of noise, especially at lower noise levels.
*   Increasing the noise strength (probability of error per qubit) significantly reduces the success rate, as the majority vote decoder can be overwhelmed by multiple errors.
*   Depolarizing noise (which includes bit-flip and phase-flip errors) tends to be slightly harder for the repetition code to correct compared to pure bit-flip noise at similar strengths, because the majority vote on Z-basis measurements primarily addresses bit-flip errors.
*   The repetition code is effective at mitigating low levels of bit-flip errors but has limitations and a noise threshold beyond which its performance degrades.

### Insights or Next Steps
